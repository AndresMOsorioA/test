4) ¬øEl continente donde est√° situado Spain es el mismo que el de `United States of America? Utiliza una condici√≥n para obtener un resultado Booleano (True o False)

```python
Paises = df["Country"]
Espanya = df[Paises == "Spain"]
USA = df[Paises == "United States of America"]
EspanyaContinente = Espanya["Continent"].values[0]
USAContinente = USA["Continent"].values[0]
bool = True
if EspanyaContinente != USAContinente:
    bool = False

print("Spain est√° situado en el continente "+ str(EspanyaContinente))
print("United States of America est√° situado en el continente "+ str(USAContinente))
print("Por ende la afirmaci√≥n de que el continente donde est√° situado Spain es el mismo que el de United States of America es "+str(bool))
```

10) Calcula la cantidad de ayuda recibida por cada municipio en funci√≥n del n√∫meto total de habitantes.

```python
import pandas as pd
import random

random.seed(0)

nombres = [f'Municipio{i}' for i in range(1, 11)]

data_municipios = {
    'Nombre': nombres,
    'C√≥digo Postal': [random.randint(10000, 99999) for _ in range(10)],
    'Poblaci√≥n': [random.randint(1000, 50000) for _ in range(10)]  # A√±adimos un atributo aleatorio, en este caso "Poblaci√≥n"
}

df_municipios = pd.DataFrame(data_municipios)


data_ayudas = {
    'Nombre': [random.choice(nombres) for _ in range(20)],
    'Ayuda Econ√≥mica (en euros)': [random.randint(1000, 10000) for _ in range(20)],
    'N√∫mero de Beneficiarios': [random.randint(10, 100) for _ in range(20)]
}

df_ayudas = pd.DataFrame(data_ayudas)

print("Dataframe de Municipios:")
print(df_municipios)

print("\nDataframe de Ayudas:")
print(df_ayudas)
```
